```python
import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```

## Download the language model - English

```python
Lang_reader = easyocr.Reader(['en'])
```

## Image Preview that will be taken for  OCR

```python
Image(r'/content/Train_because_images.jpg')
```

## Text Detection In /content/Test_because_images.jpg.jpg

```python
text_output1 = Lang_reader.readtext(r'/content/Test_because_images.jpg')
```

## Now we will print the detected texts that are present in the image

### Printing all the List Of Bounding boxes, Co-ordinates present in the image, Text and Their Confidence level

```python
text_output1
```

## Extract the text & its bounding box coordinates

```python
cord_1 = text_output1[-2][0]
print(text_output1[-2][1])
```

## Minimum value of coordinates

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord_1)]
```

## Maximum value of coordinates

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord_1)]
```

## Visuals of Bounding box co-ordinates of particular text In Image

```python
image = cv2.imread('/content/Test_because_images.jpg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
cord_2 = text_output1[-1][0]
print(text_output1[-1][1])
```
